```python
import pandas as pd

df = pd.read_csv('winequality-red.csv')

df.head()
```

```python
categorical_col = 'quality'
numeric_cols = ['alcohol', 'pH']

print("Overall Statistics:\n")
for col in numeric_cols:
    print(f"{col}:")
    print(f"  Mean: {df[col].mean():.2f}")
    print(f"  Max: {df[col].max():.2f}")
    print(f"  Min: {df[col].min():.2f}")
    print(f"  Standard Deviation: {df[col].std():.2f}")
    print(f"  Range: {df[col].max() - df[col].min():.2f}")
    print(f"  Median: {df[col].median():.2f}\n")

print("Statistics by Quality:\n")
for q in sorted(df[categorical_col].unique()):
    subset = df[df[categorical_col] == q]
    print(f"Quality {q}:")
    for col in numeric_cols:
        print(f"  {col}: Mean={subset[col].mean():.2f}, Max={subset[col].max():.2f}, Min={subset[col].min():.2f}, Std={subset[col].std():.2f}, Range={subset[col].max() - subset[col].min():.2f}, Median={subset[col].median():.2f}")
    print()
```

## Task 3: Plotting Basic Distributions
- Scatter plot: alcohol vs pH  
- Histograms: alcohol and pH  
- Histograms by category: alcohol for each wine quality

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='alcohol', y='pH', hue='quality', palette='viridis', s=60)
plt.title('Scatter Plot of Alcohol vs pH by Wine Quality')
plt.xlabel('Alcohol (%)')
plt.ylabel('pH')
plt.legend(title='Quality')
plt.show()
```

### Histograms of Numeric Columns with Mean and Median
The histograms below show the distribution of alcohol and pH values. Red dashed lines indicate the mean, and green dotted lines indicate the median.

```python

plt.figure(figsize=(8,5))
sns.histplot(df['alcohol'], bins=15, kde=False, color='skyblue')


mean_alc = df['alcohol'].mean()
median_alc = df['alcohol'].median()
plt.axvline(mean_alc, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_alc:.2f}')
plt.axvline(median_alc, color='green', linestyle='dotted', linewidth=2, label=f'Median: {median_alc:.2f}')

plt.title('Distribution of Alcohol with Mean and Median')
plt.xlabel('Alcohol (%)')
plt.ylabel('Count')
plt.legend()
plt.show()


plt.figure(figsize=(8,5))
sns.histplot(df['pH'], bins=15, kde=False, color='salmon')


mean_pH = df['pH'].mean()
median_pH = df['pH'].median()
plt.axvline(mean_pH, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_pH:.2f}')
plt.axvline(median_pH, color='green', linestyle='dotted', linewidth=2, label=f'Median: {median_pH:.2f}')

plt.title('Distribution of pH with Mean and Median')
plt.xlabel('pH')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
qualities = sorted(df['quality'].unique())
plt.figure(figsize=(12,8))

for q in qualities:
    subset = df[df['quality'] == q]
    sns.histplot(subset['alcohol'], bins=10, kde=False, label=f'Quality {q}', alpha=0.6)

plt.title('Distribution of Alcohol by Wine Quality')
plt.xlabel('Alcohol (%)')
plt.ylabel('Count')
plt.legend(title='Quality')
plt.show()
```
