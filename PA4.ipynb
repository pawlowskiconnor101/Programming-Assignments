```python
num = int(input("Please input a positive integer: "))

for i in range(1, num+1):
    if num % i == 0:
        print(f"{i} is a factor of {num}")

if num > 1:
    for i in range (2, int(num**0.5)+1):
        if num % i == 0:
            print("Your number is not prime.")
            break
    else:
        print('Your number is prime')
else: print('Your number is not prime')
```

```python
translator = {'hello': 'hola', 'cat': 'gato', 'dog': 'perro', 'chair': 'silla', 'book': 'libro','water': 'agua', 'food': 'comida', 'friend': 'amigo', 'house': 'casa', 'sun': 'sol'}

while True:
    word = input("Enter a word to translate from English to Spanish (hit enter to stop): ").lower()
    if word == "":
        break
    elif word in translator:
        print(f"The Spanish word for {word} is {translator[word]}.")
    else:
        print(f"The Spanish word for {word} is not in this dictionary.")
```

```python
temps = { 'January': [43, 27], 'February': [49, 30], 'March': [58, 38], 'April': [69, 47], 'May': [77, 57], 'June': [85, 66], 'July': [88, 70], 'August': [87, 68],
    'September': [81, 61], 'October': [70, 49], 'November': [57, 39], 'December': [46, 31]}
while True:
    month = input("Enter a month: I will give you the average high and low temperatures in Louisville\n: ").capitalize()
    if month == "Quit":
        break
    elif month in temps:
        print(f"The average high temperature in Louisville for {month} is {temps[month][0]} and the average low is {temps[month][1]}.")
    else:
        print("Not a valid month")
```

```python
inventory = { 'hammer': [15, 12.49], 'wrench': [20, 7.99], 'screwdriver': [40, 3.99], 'nailgun': [10, 36.59], 'padlock': [25, 5.99], 'roll_of_duct_tape': [30, 4.95] }

print("Menu:")
print("1. Look up an item")
print("2. Display all items below a budget")
print("3. Show total quantity of all items")
print("4. Show total value of all inventory")

choice = int(input("Enter your choice (1–4): "))

if choice == 1:
    item = input("Enter the item name: ").lower()
    if item in inventory:
        print(f"There are {inventory[item][0]} {item}s available at ${inventory[item][1]} each.")
    else:
        print("That item is not available.")
elif choice == 2:
    budget = float(input("Enter your maximum budget: "))
    print("Items below your budget:")
    for item, (qty, price) in inventory.items():
        if price < budget:
            print(f"{item}: ${price}")
elif choice == 3:
    total_qty = sum(qty for qty, _ in inventory.values())
    print(f"Total quantity of all items: {total_qty}")
elif choice == 4:
    total_value = sum(qty * price for qty, price in inventory.values())
    print(f"Total value of all inventory: ${total_value:.2f}")
else:
    print("Invalid choice.")
```
