## Task 1: Weekend Activities as Sets

```python
actSat = set()
actSun = set()

while True:
    activity = input('Enter an activity from last Saturday ("Done" to stop): ')
    if activity.strip().lower() == "done":
        break
    actSat.add(activity.strip())

while True:
    activity = input('Enter an activity from last Sunday ("Done" to stop): ')
    if activity.strip().lower() == "done":
        break
    actSun.add(activity.strip())

print("\nActivities on Saturday:", actSat)
print("Activities on Sunday:", actSun)

print("Activities on BOTH days:", actSat & actSun)

print("Activities on AT LEAST one day:", actSat | actSun)
```

## Task 2: Triangle Calculations

```python
import math

def triangle_solve(leg1, leg2):
    hyp = math.sqrt(leg1**2 + leg2**2)
    perimeter = leg1 + leg2 + hyp
    area = 0.5 * leg1 * leg2
    return (hyp, perimeter, area)

print(triangle_solve(3, 4)) 
```

## Task 3: Calculator Class

```python
class Calculator:
    
    def __init__(self):
        self.prev_result = None
        self.total_calculations = 0
    
    def add(self, a, b):
        result = a + b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def subtract(self, a, b):
        result = a - b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def multiply(self, a, b):
        result = a * b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def divide(self, a, b):
        if b == 0:
            raise ZeroDivisionError("Cannot divide by zero.")
        result = a / b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def power(self, a, b):
        result = a ** b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def info(self):
        print(f"Previous Result: {self.prev_result}")
        print(f"Total Calculations: {self.total_calculations}")


calc = Calculator()
print(calc.add(5, 3))
print(calc.multiply(2, 4))
calc.info()
```
